```python
import pandas as pd
import numpy as np
```

```python
best= pd.read_csv('сабмиссия лучшая по первому разделению_115414.csv', sep=',', index_col=None)
cand_22= pd.read_csv('тест 22 неизвестных.csv', sep=',', index_col=None)
cand_3= pd.read_csv('тест 3 наиболее значимых.csv', sep=',', index_col=None)
test= pd.read_csv('test_predobr.csv', sep=';', index_col=None)
claster_for3=pd.read_csv('tabliza_clasters_for3.csv', sep=';', index_col=None)
```

```python
# лучшая на момент разделения по верхней части сабмиссия
best['overall_worklogs'].value_counts()
```

```python
# Найденные по результатам модели разделения наиболее значимые записи (с результатами, присвоенными перебором по 10 кластерам
# значений трейна от 68000)
cand_3
```

```python
# Таблица кластеров, по которой проводилось распределение значений по 3-м
claster_for3
```

```python
# Найденные по модели второго разделния 22 значения без изменений паблика при изменении значений
# (вероятно, записи для проверки модели на привате)
cand_22.merge(best, how="left", on="id")
```

```python
# Создание теста для модели прогнозирования значений 22 неизвестных
test_22=cand_22.merge(test, how="left", on="id")
test_22.to_csv('test_22.csv', sep=';', index=None)
test_22
```

```python
# из 3-х точно найденных больших значений формирует контрольный пул (проверка на точность предсказаний модели)
control_pool=cand_3[['id']]
control_test=control_pool.merge(test, how="left", on="id")
control_test.to_csv('control_test.csv', sep=';', index=None)
control_test
```

```python
# Загрузка таблицы 5 прогнозов, сделанных последней моделью для 22 неизвестных (с указанием сидов модели для воспроизведения)
prognoses_22= pd.read_csv('prognoses for 22.csv', sep=',', index_col=None)
prognoses_22
```

```python
# Формируем усреднение 5-и прогнозов.
# С целью минимизации потерь при ошибке, наибольшие значения фиксируем по центру кластера нижней границы
prognoses_22['Abs']=np.round((prognoses_22['Sid 8']+prognoses_22['Sid 21']+prognoses_22['Sid32']+prognoses_22['Sid43']+prognoses_22['Sid77'])/5, 0)
prognoses_22.loc[(prognoses_22['Abs']>262089),'Abs']=262089
prognoses_22.loc[(prognoses_22['Abs']>136290)&(prognoses_22['Abs']<262089),'Abs']=136290
prognoses_22
```

```python
# Сводим прогноз по 22 и прогноз по 3 в одну группу
prognoses_22=prognoses_22[['id','Abs']]
prognoses_22.columns=['id','overall_worklogs']
prognoses_25=pd.concat([prognoses_22,cand_3])
prognoses_25['overall_worklogs']=prognoses_25['overall_worklogs'].astype(int)
```

```python
# Вводим прогнозы в лучший исходный вариант
prognoses_25.columns=['id','var1']
new=best.merge(prognoses_25, how="left", on="id")
new.loc[(~new['var1'].isna()),'overall_worklogs']=new['var1']
new=new[['id','overall_worklogs']]
new['overall_worklogs']=new['overall_worklogs'].astype(int)
new['overall_worklogs'].value_counts()
```

```python
new.to_csv('итоговый прогноз.csv', sep=',', index=None)
```
